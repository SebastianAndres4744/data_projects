In this Project we will analyze the Iris Dataset & implement a Classification Model that will classify the type of Flower for each record within the dataset.

This dataset contains multiple characteristics of a total of 150 Iris Plants:
* sepal lenght (cm)
* sepal width (cm)
* petal lenght (cm)
* petal width (cm)

The Iris Plants can be classified in 3 different types:
* Iris Setosa
* Iris Versicolour
* Iris Virginicas

With these atributes, our Model will be able to analyze the characteristics of a plant and classify it under one of the three diferent types of Iris Plants.

### 1. Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
import seaborn as sns
```

```python
iris = datasets.load_iris()
```

### Visualizing the Total No. of Cases per Iris Plant Type - Barchar

```python
# We count the occurrences of each species in the target array:
unique, counts = np.unique(iris.target, return_counts = True)

# We get the Species Names:
species_names = [iris.target_names[i] for i in unique]

# We then create a barchar:
plt.bar(species_names, counts, color = ['red', 'green', 'blue'])

# Add labels & title:
plt.xlabel('Iris Plant Species')
plt.ylabel('Number of Cases')
plt.title('Total Number of Cases per Iris Plant Species in Dataset')

# Display the Bar:
plt.show()
```

### Visualize the Distribution of Iris Dataset Features

```python
# Convert the dataset into a pandas DataFrame
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# Add the species column to the DataFrame
df_iris['species'] = iris.target
df_iris['species'] = df_iris['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# Set up the style for better visualization
sns.set_style("whitegrid")

# Create histograms to see the distribution of each feature
df_iris.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle('Distribution of Iris Dataset Features', fontsize=16)
plt.show()

# Create boxplots to observe the feature distribution per species
plt.figure(figsize=(12, 8))
for i, column in enumerate(df_iris.columns[:-1]):  # Exclude the 'species' column
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='species', y=column, data=df_iris, palette='Set2')
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

## Analysis and Insights from the Histograms of the Iris Dataset Features:
### 1. Sepal Length Distribution (Top-left chart)

##### Observations:

The sepal length ranges between approximately 4.5 cm to 8.0 cm.

The distribution appears to be multimodal, with peaks around 5.0 cm, 5.8 cm, and 6.5 cm.

There is a relatively even spread across the values, but certain lengths (e.g., around 5.5 cm) occur more frequently.

##### Insights:

The distribution suggests that different species may have distinct sepal length characteristics.
A potential clustering of values could indicate species differentiation based on sepal length.

### 2. Sepal Width Distribution (Top-right chart)

##### Observations:

The sepal width ranges between approximately 2.0 cm to 4.5 cm.

The distribution is slightly right-skewed, with a peak around 3.0 cm.

There are fewer instances of very low or very high values.

##### Insights:

The peak suggests that most flowers have a sepal width around 3.0 cm.

The right skewness indicates that fewer flowers have significantly wider sepals.

### 3. Petal Length Distribution (Bottom-left chart)

##### Observations:

The petal length has a bimodal distribution with peaks around 1.5 cm and 4.5 cm.

There is a clear distinction between short petals (~1.5 cm) and longer petals (~4.5 cm to 5.5 cm).

The dataset exhibits two major groups, suggesting a strong differentiation in petal length.

##### Insights:

The bimodal distribution likely indicates different species with distinct petal length characteristics.

Setosa species might be contributing to the lower values, while Versicolor and Virginica to the higher values.

### 4. Petal Width Distribution (Bottom-right chart)

##### Observations:

The petal width distribution is also bimodal, with peaks around 0.2 cm and 1.5 cm.

There is a distinct gap between the two peaks, showing two major groups in the data.

A significant portion of the dataset has very small petal widths.

##### Insights:

The clear separation suggests that petal width is a strong differentiator among the Iris species.

Setosa likely has the smaller petal widths, while the other species exhibit wider petals.

Overall Insights from the Iris Dataset Distributions:

The features petal length and petal width show clear bimodal distributions, making them strong indicators for distinguishing species.

Sepal length and sepal width are more normally distributed, suggesting that they may have less discriminative power compared to petal features.

The distributions suggest that clustering or classification algorithms will likely perform well due to the clear separations in petal-based features.

### 6. Assigning Output & Input Variables:

We'll assing the 4 input features to the X variable & the output target to the Y variable:


```python
X = iris.data
y = iris.target
```

## 7. Build a Classification Model using Random Forest

```python
random_clf = RandomForestClassifier()
```

```python
random_clf.fit(X, y)
```

## 8. Feature Importance

Let's examine the importance that our Random Forest Classifier assigned to each Feature of our Iris Dataset

```python
# Combione features names with their importance scores:
for feature, importance in zip(iris.feature_names, random_clf.feature_importances_):
    print(f"{feature}: {importance:.4f}")
```

As we can see, the two most important features for the Random Forest Classifier are 
**Petal Lenght (cm)**
**Petal Width (cm)**

The feature importance values obtained from the random forest classifier align well with the insights drawn from the histograms of the feature distributions.

## Comparison and Interpretation:
### Petal Length (0.4702 - Most Important Feature)

#### Histogram Insight:

The histogram of petal length showed a bimodal distribution, indicating a clear separation between species.

The significant gap between the two groups suggests that petal length is a strong differentiator among species.

* Interpretation:

The high importance score confirms that petal length is indeed the most crucial feature in classifying Iris species, which aligns with the visual evidence of distinct clusters in the histogram.

### Petal Width (0.3840 - Second Most Important Feature)

#### Histogram Insight:
The petal width distribution also exhibited a bimodal pattern, with a clear distinction between smaller and larger values.

The separation of clusters indicated strong species differentiation.
* Interpretation:
The high feature importance confirms that petal width is another key predictor of species, reinforcing the insight derived from the histogram.

### Sepal Length (0.1209 - Less Important Feature)

#### Histogram Insight:

Sepal length had a more evenly distributed and multimodal pattern, suggesting less clear separability between species.

* Interpretation:

Its lower feature importance score reflects that it provides some value but is not as decisive as petal-based features in classifying the species.

### Sepal Width (0.0249 - Least Important Feature)

## 10. Doing a Train Test Split on our Dataset (80/20)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2)
```

We assign 80% of the Dataset to the Training Set & 20% to the Testing Set

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

## 11. Rebuild the Random Forest Model with the Training Set only

```python
random_clf.fit(X_train, y_train)
```

## 12. Perform Prediction on Test Set & evaluating Model Performance

```python
yhat = random_clf.predict(X_test)
```

### 12.1 Accuracy Score

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test, yhat)
print(f'Accuracy: {accuracy:.4f}')
```

### 12.2 Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, yhat)
print(cm)
```

### 12.3 Classification Report

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, yhat)
print(report)
```

Interpretation by Class:

Precision:

Recall (Sensitivity):

Recall measures how many actual positive cases were correctly identified.
* For class 0 (Setosa): Recall = 1.00, meaning all Setosa instances were correctly classified.
* For class 1 (Versicolor): Recall = 0.89, meaning some Versicolor instances were missed.
* For class 2 (Virginica): Recall = 1.00, meaning all Virginica samples were correctly classified.

F1-score:

The F1-score is the harmonic mean of precision and recall, providing a balanced measure.

High F1-scores for all classes indicate balanced performance.

Support:

Represents the number of actual occurrences for each class in the test set.

Overall Performance:


Macro average: (arithmetic mean of precision, recall, and F1-score across classes) = 0.97, meaning the model performs consistently across classes.

Weighted average: (accounts for the number of instances in each class) = 0.97, indicating overall model effectiveness.
